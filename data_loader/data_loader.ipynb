```python
from raw_data import glyphs
import chromadb
from tqdm import tqdm

client = chromadb.Client()
collection = client.create_collection(name="glyphs")
glyphs
```

# extract data

```python
def parse_glyph_name(name:str)->tuple[str,str, str]:
    """
    Args:
        glyph_name, such as "nf-cod-arrow_small_left"
    
    Return:
        series, group, description
    """
    series, group, description_text = name.split("-")
    return series, group, description_text.replace("_", " ")
```

```python
datas = []
for glyph_name, unicode in glyphs.items():
    series, group, description = parse_glyph_name(glyph_name)
    if group != "mdi":
        datas.append({
            "ids": glyph_name,
            "metadatas": {
                "series": series,
                "group": group,
                "unicode": unicode
            },
            "documents": description,
        })

datas
```

# load data

```python
for data in tqdm(datas):
    collection.add(**data) 
```

```python
collection.query(query_texts="md cat", n_results=10)
```

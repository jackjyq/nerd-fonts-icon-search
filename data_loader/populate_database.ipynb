```python
import chromadb
client = chromadb.Client()
```

```python
collection = client.create_collection(name="glyphs")
```

```python
collection.add(
    documents=[key for key in glyphs],
    ids=[key for key in glyphs],
    metadatas=[{"unicode": value} for value in glyphs.values()]
)
```
